```python
import pandas as pd
import numpy as np
```

```python
laptops_clean   = pd.read_csv("/content/laptops_cleaned.csv")
laptops_unclean = pd.read_csv("/content/laptops_uncleaned.csv")
ratings_df      = pd.read_csv("/content/personalized_recommendation_dataset.csv")
```

```python
print("=" * 60)
print("LAPTOPS — CLEANED VERSION")
print("=" * 60)
print(f"Rows: {laptops_clean.shape[0]}   Columns: {laptops_clean.shape[1]}")
print("\nColumn names and types:")
print(laptops_clean.dtypes)
print("\nFirst 5 rows:")
print(laptops_clean.head())
```

```python
print("\n" + "=" * 60)
print("LAPTOPS — UNCLEANED VERSION")
print("=" * 60)
print(f"Rows: {laptops_unclean.shape[0]}   Columns: {laptops_unclean.shape[1]}")
print("\nColumn names and types:")
print(laptops_unclean.dtypes)
print("\nFirst 5 rows:")
print(laptops_unclean.head())
```

```python
print("\n" + "=" * 60)
print("MISSING VALUES — LAPTOPS CLEANED")
print("=" * 60)
missing_clean = laptops_clean.isnull().sum()
missing_clean_pct = (missing_clean / len(laptops_clean) * 100).round(1)
missing_clean_df = pd.DataFrame({
    "Missing Count": missing_clean,
    "Missing %": missing_clean_pct
})
print(missing_clean_df[missing_clean_df["Missing Count"] > 0])
```

```python

print("\n" + "=" * 60)
print("MISSING VALUES — LAPTOPS UNCLEANED")
print("=" * 60)
missing_unclean = laptops_unclean.isnull().sum()
missing_unclean_pct = (missing_unclean / len(laptops_unclean) * 100).round(1)
missing_unclean_df = pd.DataFrame({
    "Missing Count": missing_unclean,
    "Missing %": missing_unclean_pct
})
print(missing_unclean_df[missing_unclean_df["Missing Count"] > 0])
```
